````python
import matplotlib.pyplot as plt
import pandas as pd

# Load the financial data from CSV
df = pd.read_csv('fdata.csv')


# Convert 'Date' column to datetime format and format as 'MMM DD YYYY', ('%d\n%b\n%Y') then day in digit \n month in words \n year in digit
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y').dt.strftime('%d\n%b\n%Y')

# Extract data for plotting
dates = df['Date']
open = df['Open']
high = df['High']
low = df['Low']
close = df['Close']

# Plotting each attribute
plt.plot(dates, open, linestyle='-', label='Open')
plt.plot(dates, high, linestyle='-', label='High')
plt.plot(dates, low, linestyle='-', label='Low')
plt.plot(dates, close, linestyle='-', label='Close')


# Add a legend
plt.legend()

# Display the plot
plt.tight_layout()
plt.show()
````

##6.  Write a Python programming to display a bar chart of the popularity of programming Languages.
```
Sample data:
Programming languages: Java, Python, PHP, JavaScript, C#, C++
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
The code snippet gives the output shown in the following screenshot:
```

image.png

````python
import matplotlib.pyplot as plt

# Data for plotting
programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

# Set size
plt.figure(figsize=(8, 6))  # Adjust figure size (width, height)

# Plotting the bar chart
plt.bar(programming_languages, popularity, color='blue',align='edge')

# Set axis limit
plt.ylim(bottom = 0, top=25)
plt.xlim(left = 0)

# Adding labels and title
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\nWorldwide, Oct 2017 compared to a year ago')

# Main grid red
plt.grid(True, which= 'major',color='red',linewidth=0.5)

# Subgrid with dotted lines
plt.minorticks_on() # Enable minor ticks on the plot
plt.grid(which='minor', color='black', linestyle=':', linewidth=0.5)

plt.tight_layout()  # Adjust layout for better appearance

# Display the plot
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('company_sales_data.csv')

# Extract month_number , total_profit columns
month_number = data['month_number']
total_profit =  data['total_profit']
total_profit_of_all_months = data['total_profit'].sum()

# Set figure size (width, height)
plt.figure(figsize=(9, 7))

# Create a line plot with specified style properties
# Line Style dotted and Line-color should be red
# Show legend at the lower right location.
# X label name = Month Number
# Y label name = Sold units profit
# Add a circle marker.
# Line marker color as read
# Line width should be 3
plt.plot(month_number, total_profit, linestyle='--', color='red', marker='o', markerfacecolor='black', markersize=10, linewidth=3, label='Profit data of last year')

# Set the labels and title
plt.xlabel('Month Number')
plt.ylabel('Sold units profit')
plt.title('Company Sales data of last year')
# Annotate the total profit of all months on the graph
plt.text(0.89,0.09, f'Total Profit: {total_profit_of_all_months}', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes, fontsize=10, bbox=dict(facecolor='white', alpha=0.1))

# The transform=plt.gca().transAxes argument ensures the text is placed in the coordinate space of the plot (0.5, 0.5) represents the center of the plot.
# The bbox argument is used to add a background color with some transparency to the text box for better visibility.

# Show legend at the lower right location
plt.legend(loc='lower right')

plt.tight_layout()  # Adjust layout for better appearance

# Display the plot
plt.show()
````

##Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.



image.png

````python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('company_sales_data.csv')

# Extract month_number , sales units columns and calculate the total sales for each product
month_number = data['month_number']
facecream = data['facecream'].sum()
facewash = data['facewash'].sum()
toothpaste = data['toothpaste'].sum()
bathingsoap = data['bathingsoap'].sum()
shampoo = data['shampoo'].sum()
moisturizer = data['moisturizer'].sum()

total_sale_data = [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']

# Calculate the percentage of each product's sales
percent = [total / sum(total_sale_data) * 100 for total in total_sale_data]

# Set size
plt.figure(figsize=(10, 10))  # Adjust figure size (width, height)

# Plot pie chart
# autopct='%1.1f%%' means that the percentages will be displayed with one decimal place and a percent sign.
# 1.1f: This specifies the format of the number:
# Using %% escapes the percent sign so it is displayed as a literal character in the output.
plt.pie(percent, labels=labels, autopct='%1.1f%%', startangle=0, colors=['#357AAF','#F1843E','#50993A','#C73D3A','#8D71B9','#86584F'])

# Add title
plt.title('Number of units sold per year in percentage.')

# Show legend at lower right location
plt.legend(loc='lower right')

# Show the pie chart
plt.show()
````

##Exercise Question 10: Read all product sales data and show it using the stack plot
The Stack plot should look like this.

image.png


````python
# here we are using the same colors as per the pie chart

import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('company_sales_data.csv')

# Extract and define plot data
month_number = data['month_number']
facecream = data['facecream']
facewash = data['facewash']
toothpaste = data['toothpaste']
bathingsoap = data['bathingsoap']
shampoo = data['shampoo']
moisturizer = data['moisturizer']

sale_data = [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
labels = ['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
colors = ['#357AAF','#F1843E','#50993A','#C73D3A','#8D71B9','#86584F']

# Set figure size
plt.figure(figsize=(10, 6))

# Create a stacked plot
plt.stackplot(month_number,sale_data,labels=labels,colors=colors)

# Add labels and title
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('All Product Sales Data')

# Show legend
plt.legend(loc='upper left')

# Show the plot
plt.tight_layout()
plt.show()
````
